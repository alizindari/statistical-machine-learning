```python
import numpy as np
import matplotlib.pyplot as plt

from scipy import stats


def plot_data(x, t):
    plt.scatter(x, t, marker='o', c="k", s=20)


def plot_truth(x, y, label='Truth'):
    plt.plot(x, y, 'k--', label=label)


def plot_predictive(x, y, std, y_label='Prediction', std_label='Uncertainty', plot_xy_labels=True):
    y = y.ravel()
    std = std.ravel()

    plt.plot(x, y, label=y_label)
    plt.fill_between(x.ravel(), y + std, y - std, alpha = 0.5, label=std_label)

    if plot_xy_labels:
        plt.xlabel('x')
        plt.ylabel('y')


def plot_posterior_samples(x, ys, plot_xy_labels=True):
    plt.plot(x, ys[:, 0], 'r-', alpha=0.5, label='Post. samples')
    for i in range(1, ys.shape[1]):
        plt.plot(x, ys[:, i], 'r-', alpha=0.5)

    if plot_xy_labels:
        plt.xlabel('x')
        plt.ylabel('y')





def print_comparison(title, a, b, a_prefix='np', b_prefix='br'):
    print(title)
    print('-' * len(title))
    print(f'{a_prefix}:', a)
    print(f'{b_prefix}:', b)
```

```python
rng = np.random.RandomState(1)
biases = np.ones(50)
x = 10 * rng.rand(50)
x_train = np.concatenate([biases.reshape(-1,1),x.reshape(-1,1)],axis=1)
y = x+rng.randn(50)
plt.scatter(x, y)
```

```python
theta = (np.linalg.inv((x_train.T)@x_train))@((x_train.T)@y)
```

```python
theta
```

```python
xfit = np.linspace(0, 10, 1000)
yfit = theta[0]+theta[1]*xfit
plt.scatter(x, y)
plt.plot(xfit, yfit,'ro',markersize=0.5)
```

```python
rng = np.random.RandomState(1)
x = 10 * rng.rand(200)
y = np.sin(x) + 0.1 * rng.randn(200)

plt.scatter(x, y)
```

```python
x_train = x.reshape(-1,1).copy()
ones = np.ones(200).reshape(-1,1)
feature1 = x_train.copy()**2
feature2 = x_train.copy()**3
feature3 = x_train.copy()**4
feature4 = x_train.copy()**5
feature5 = feature2+feature3
```

```python
X = np.concatenate([ones,feature1,feature2,feature3,feature4,feature5],axis=1)
```

```python
theta = (np.linalg.inv((X.T)@X))@((X.T)@y)
```

```python
xfit = np.linspace(0, 10, 800)
yfit = theta[0]+theta[1]*(xfit**2)+theta[2]*(xfit**3)+theta[3]*(xfit**4)+theta[4]*(xfit**5)+theta[5]*(xfit**3+xfit**4)
```

```python
plt.scatter(x, y)
plt.plot(xfit, yfit,'ro',markersize=0.5);
```

```python
k = 50
x = np.concatenate([np.ones(k-1).reshape(-1,1),np.arange(1,k).reshape(-1,1)],axis=1)
y = (7*x[:,1]+34).reshape(-1,1)
#################################

```

```python
def posterior(x,y,beta,pr_mean,pr_cov):
    pos_cov = np.linalg.inv(pr_cov+beta*(x.T@x))
    temp = np.linalg.inv(pr_cov)@pr_mean+beta*(x.T@y)
    pos_mean = pos_cov@temp
    return pos_mean,pos_cov
```

```python
beta = 2.0
alpha = 1.0
pr_mean = np.zeros(x.shape[1]).reshape(2,1)
pr_cov = alpha*np.eye(x.shape[1])

posterier_mean,posterior_covariance = posterior(x,y,beta,pr_mean,pr_cov)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
 
 
# plt.style.use('seaborn-dark')
# plt.rcParams['figure.figsize']=14,6
# fig = plt.figure()
 
# Initializing the random seed
random_seed=1000
 
# List containing the variance
# covariance values
cov_val = [-0.05, 0, 0.05]
 
# Setting mean of the distributino
# to be at (0,0)
mean = np.array([20,20])
 
# Storing density function values for
# further analysis
pdf_list = []
 
# Iterating over different covariance values
for idx, val in enumerate(cov_val):
     
    # Initializing the covariance matrix
    cov = np.array([[1, val], [val, 1]])
     
    # Generating a Gaussian bivariate distribution
    # with given mean and covariance matrix
    distr = multivariate_normal(cov = cov, mean = mean,
                                seed = random_seed)
     
    # Generating a meshgrid complacent with
    # the 3-sigma boundary
    mean_1, mean_2 = mean[0], mean[1]
    sigma_1, sigma_2 = cov[0,0], cov[1,1]
     
    x = np.linspace(-25*sigma_1, 25*sigma_1, num=100)
    y = np.linspace(-25*sigma_2, 25*sigma_2, num=100)
    X, Y = np.meshgrid(x,y)
     
    # Generating the density function
    # for each point in the meshgrid
    pdf = np.zeros(X.shape)
    for i in range(X.shape[0]):
        for j in range(X.shape[1]):
            pdf[i,j] = distr.pdf([X[i,j], Y[i,j]])
     
    # Plotting the density function values

    pdf_list.append(pdf)
    


for idx, val in enumerate(pdf_list):
    plt.subplot(1,3,idx+1)
    plt.contourf(X, Y, val, cmap='viridis')
    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.title(f'Covariance between x1 and x2 = {cov_val[idx]}')
# plt.tight_layout()
plt.show()
```
