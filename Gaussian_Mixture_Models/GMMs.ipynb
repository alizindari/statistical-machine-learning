```python
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import scipy.stats
from sklearn.datasets import make_spd_matrix
import pdb
import random
```

```python
x,y = make_blobs(n_samples=500, centers=2, n_features=2)
plt.scatter(x[:,0],x[:,1])
```

```python
m_new1 = np.array([1,2],dtype=np.float32)
m_new2 = np.array([9,7],dtype=np.float32)

new_cov1 = make_spd_matrix(2,2).astype(np.float32)
new_cov2 = make_spd_matrix(2,2).astype(np.float32)

pi1 = 0.1
pi2 = 0.9

norm1 = scipy.stats.multivariate_normal(m_new1,new_cov1)
norm2 = scipy.stats.multivariate_normal(m_new2,new_cov2)
```

```python
for it in range(100):
    respons = []
#     print('d')
    norm1 = scipy.stats.multivariate_normal(m_new1,new_cov1)
    norm2 = scipy.stats.multivariate_normal(m_new2,new_cov2)
    
    ##################################################
    #            Computing New Responsibilities      #
    ##################################################
    
    for i in range(x.shape[0]):
        r1 = pi1*norm1.pdf(x[i,:])/(pi1*norm1.pdf(x[i,:])+pi2*norm2.pdf(x[i,:]))
        r2 = pi2*norm2.pdf(x[i,:])/(pi1*norm1.pdf(x[i,:])+pi2*norm2.pdf(x[i,:]))
        respons.append([r1,r2])
    respons = np.array(respons)
    
    ##################################################
    #            Computing New Mean Vectors          #
    ##################################################
    
    temp1 = np.array([0,0],dtype=np.float32)
    temp2 = np.array([0,0],dtype=np.float32)
    
    for i in range(x.shape[0]):
        temp1 += respons[i,0]*x[i,:]
        temp2 += respons[i,1]*x[i,:]
    m_new1 = temp1/np.sum(respons[:,0]).copy()
    m_new2 = temp2/np.sum(respons[:,1]).copy()
    
    ##################################################
    #          Computing New Covariance Matrices     #
    ##################################################
    
    temp1 = np.array([[0,0],[0,0]],dtype=np.float32)
    temp2 = np.array([[0,0],[0,0]],dtype=np.float32)
    
    for i in range(x.shape[0]):
        temp1 += respons[i,0]*(x[i,:].reshape(-1,1)-m_new1.reshape(-1,1))@((x[i,:].reshape(-1,1)-m_new1.reshape(-1,1)).T)
        temp2 += respons[i,1]*(x[i,:].reshape(-1,1)-m_new2.reshape(-1,1))@((x[i,:].reshape(-1,1)-m_new2.reshape(-1,1)).T)
     
    new_cov1 = temp1/np.sum(respons[:,0]).copy()
    new_cov2 = temp2/np.sum(respons[:,1]).copy()
    
    pi1 = np.sum(respons[:,0])/x.shape[0]
    pi2 = np.sum(respons[:,1])/x.shape[0]
```

```python
plt.scatter(x[:,0],x[:,1])
plt.scatter(m_new1[0],m_new1[1],c='r',s=122)
plt.scatter(m_new2[0],m_new2[1],c='r',s=122)
```

```python
x1 = np.linspace(-10,12,200)  
x2 = np.linspace(-10,12,200)
X, Y = np.meshgrid(x1,x2) 

Z1 = scipy.stats.multivariate_normal(m_new1, new_cov1)  
Z2 = scipy.stats.multivariate_normal(m_new2, new_cov2)

pos = np.empty(X.shape + (2,))                
pos[:, :, 0] = X; pos[:, :, 1] = Y   

plt.figure(figsize=(10,10))                                                          
plt.scatter(x[:,0], x[:,1], marker='*')     
plt.contour(X, Y, Z1.pdf(pos), colors="r" ,alpha = 1) 
plt.contour(X, Y, Z2.pdf(pos), colors="r" ,alpha = 1) 
plt.axis('equal')                                                                
plt.xlabel('X-Axis', fontsize=16)                                               
plt.ylabel('Y-Axis', fontsize=16)                                               
plt.title('Initial State', fontsize=22)                                         
plt.grid()                                                                      
plt.show()
```
